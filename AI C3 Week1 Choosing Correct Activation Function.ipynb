### Activator functions

Recall that activator functions are used in neural networks and are the sigmoid activation functions used in logistic regression, so it is called the logistic sigmoid function:

$$\hat y=a_3=f(z_3)$$

In the feed forward neural networks we looked at, such a function was applied twice.  This function is:

$$\Phi(z)=\dfrac{1}{1+e^{-z}}
$$



Another activation function that is widely used is hyperbolic tangent, or tanh. Tanh has the advantage that it also covers a negative range.



Tanh can also be used in the output layer of a binary classification model.

In the case of a multiclass specification, the output layer uses softmix.  Softmix is defined as a function which takes the value of z and then takes e to the power of it.

$$\sigma: \mathbb{R}^K \rightarrow [0,1]^K \\
\sigma(\mathbf z)_j=\dfrac{e^{z_j}}{\sum_{k=1}^K e^{z_k}} for j=1,\dots, K
$$

This softmix function has the important property that all of it's classes probabilities sum to 1.  The class you have predicted will sum close to one and the classes you did not predict will some close to zero.  So this function fits with one hot encoded vectors in the output layer.

Another activation function is called ReLU:



ReLU doesn't look non-linear but it is.  And it turns out that when a combination of multiple ReLU functions you can approximate any mathematical function.  ReLU is the most used activation function today.  This is first of all since computational complexity is less and also because networks tend to converge faster using ReLU.

One problem with ReLU is that it can cause dead neurons because a range value ouput is zero.   Therefore there is a variation of ReLU called leaky ReLU:



On the negative side of leaky ReLU there is a little bit of negative value.  This helps to prevent the problem with dead neurons.  There is also a variation called random ReLU:



It is best to start with ReLU for the input and the hidden layers.  And for the output layers, it depends on your task. If you are using a linear regressor, you use a linear output unit, and if you are creating a classifier, then you use softmix or sigmoid for the output layer.
