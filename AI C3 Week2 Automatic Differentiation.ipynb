### Automatic Differentiation

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

x = np.arange(0, 5, 0.1)
cos = np.cos(x)
# the derivative of cos(x) is -sin(x)
sin = -np.sin(x)

plt.plot(x,sin)
plt.plot(x,cos)

# the blue line is the plot of cos(x)
# the red line is the plot of cos'(x) = -sin(x)
```

```python
# this code makes sure that we are using tensorflow 1.X and not tensorflow 2.0
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

# here we start with a value x and apply Tensorflow's cosine operation on it:
x=tf.Variable(initial_value=0.3)
y=tf.cos(x)
```

```python
# this tells tensorflow to minimize over y
# in order to achieve this, tensorflow must calculate the
# first derivative of the cosine function
# (which allows it to find the minimum of the cosine function)

train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(y)
```

```python
with tf.Session() as sess:
    writer=tf.summary.FileWriter('Logs', sess.graph)
    writer.close()
```

Tensorflow is capable of automatically calculating the derivative of
any function defined in a tensorflow graph
