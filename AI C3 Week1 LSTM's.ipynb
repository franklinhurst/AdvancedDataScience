### Long Short Term Memory Networks (LSTMs)

This is a diagram is of a single neuron in an LSTM.  As is also the case with the feed forward network, it maps an input vector $x_t$ to an output vector $h_t$ by using weights and an activation function.  Note that the same holds whether you are using scalars or vectors as inputs or outputs.  But you can see there there are a number of additional components.  First, there is no direct connection between $x_t$ and $h_t$.  All information flows through $c_t$ which is the cell state.  The cell state is the memory of the LSTM neuron.  There are three additional units present: an input gate, and output gate, and a forget gate.  Those three gates are controlling the state of $c_t$.  

The way that $c_t$ is controlled is as follows.  $x_t$ is not only used as input to the neuron, but also as input to the intput gate $i_t$.  

The input gate has a separate weight vector which is straight from the input data, and learns to control the influx of information into the cell state $c_t$.  The control occurs be taking the vector dot product between the input $x_t$ after it has been squashed by the activation function and the output of the input gate.  In other words, through the weight vector of the intput gate the neuron learns from creating data when it is a good idea to open the gate  and have the input stored in the cell or when it is a bad idea to remember things and so close the influx of new information into the cell state.  Note that this is a continuous value (meaning how good of an idea it is to let new information in is stored as a continuous value).  This letting new data in is like a door that can be partially open or closed.  Finally it is important to note that the cell state has an influence on the input gate.  This is also accomplished by a separate weight vector so that the actual input gate is controlled by the historic cell state as well as by the actual value of $x_t$.

Now let's look at the output gate.  The output gate is also controlled by the actual value of %x_t% and by the cell state $c_t$.  The output gate controls how much of cell state $c_t$ gets output to down stimulants connected to $h_t$.

This topology is the initial LSTM proposed by S Hochreiter and J Schmidhuber in 1997.  In 1999 Felix Gers, Schmidhuber and Fred Cummins added an additional part, the forget gate, $f_t$.  They discovered that without the capability of forgetting the cell state $c_t$ may grow indefinitely and eventually causes the network to break down.  Again the forget gate is controlled by the input $x_t$ and the current cell state $c_t$.  And again through calculation of a dot product between the output of the forget gate and the previous cell state $c_t$ is controlled out much of the previous cell state $c_t$ is preserved.
