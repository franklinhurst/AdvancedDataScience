```python
import torch
from PIL import Image
import numpy as np
```

```python
# we create a simple tensor of zeros with size 10
torch.zeros(10)
```

```python
# to create an order two tensor, give it two arguments:
torch.zeros(2,3)
```

```python
torch.ones(2,3)
```

```python
# can also give a list of shapes as an argument to create a tensor
# this does exactly the same thing as the command above
torch.ones([2,3])
```

```python
# create a tensor and place it in a variable and then call shape and size on it
v = torch.ones(5)
print(v.size())
```

```python
# you can take the size along each of the dimensions
v = torch.ones(5)
print(v.size(0))
```

```python
# you can take the size along each of the dimensions
v = torch.ones(5,10,15)
print(v.size(1))
```

```python
# if you want to change the shape of a tensor without changing any elements in the tensor, use view
v.view((15,10,5))
```

```python
# if you want to change the shape of a tensor without changing any elements in the tensor, use view
v.view(750)
# 15(10)(5) = 750 
```

```python
# if you want to flatten out a tensor just call view with negative one: 
v.view(-1)
```

```python
# And what the negative one in the view will do is it computes how many elements need to be
# along the negative one dimension. So if we would call view with two comma negative one,
# it will automatically figure out in order to reshape the tensor
# into something that's two by x, x needs to be five.
```
