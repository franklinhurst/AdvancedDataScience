### Bias Variance Tradeoff

You want to prevent overfitting of a neural network, just like you want to prevent overfitting of any machine learning model.  A neural network is a universal function approximator.  On the one hand, this is very powerful.  But on the other hand, it means that it tends to fit any dataset ideally that that can be a problem, sbecause any dataset contains a lot of noise and the noise does not always reflect the truth of the underlying physical system.  Therefore, you don't want to fit your neural network too strongly to your data.  You want it to be more generalized.  Therefore you divide your data into a training and test set.  You train on the training dataset.  After training, you assess the performance.  If performance is equally good, you are fine.  If performance drops when testing using the test data, then you probably have an overfittting problem.  Here we see a 2D non-linear separation.



If the training of the model draws the line "too well" then it will not be general enough to apply to other datasets, which is the ultimate purpose of a model.  If overfitted, the model has very high accuracy on the training dataset.

How can we prevent overfitting in a neural network?

1.  One way is to just get more data, and use more data to continue to train the model so that it is more general.  As a rule of thumb, you should have 10 times more data than you have parameters (weights) in your model.

2.  Regularization: In the cost function, you penalize very high weights.  You just penalize the high weights and you make those high weights cost more in the cost function.  Since your network is trying to minimize the cost function, then your model will train itself in a way that avoids the high weights.

3.  Early stopping:  This means that you don't train until you have full convergence with the training set.  You stop before that.  This should prevent overfitting.  You stop at the number of epochs where the validation data has the lowest error, not at the point where the training data has its lowest error:



4.  Drop Out:  You randomly deactivate neurons in each epoch or in each iteration.  So this causes the neural network to take other paths through the graph during training and therefore helps with generalization.
